```python
# Funcion para comprobar la integridad de los datos de entrada
def comprobarDatosEntrada(numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas):
    todoOK = False
    mensaje = ""
    
    if (type(numVacas) == int):
        if type(pesoMaxCamion) == int:
            if (len(listaPesosVacas)) == numVacas:
                if (len(listaLitrosVacas)) == numVacas:
                    todoOK =True
                else:
                    mensaje = "Las lista de litros por vaca no coincide con el nº de vacas indicado"
            else:
                mensaje = "Las lista de pesos de las vacas no coincide con el nº de vacas indicado"
        else:
            mensaje = "El peso máximo del camión debe ser un numero sin decimales"
    else:
        mensaje = "El numero de vacas debe ser un numero sin decimales"
    
    return todoOK, mensaje
```

```python
# Funcion que crea la tabla que llevará la información de las vacas y otra que recogera los posibles maximos beneficios
def crearMatrices (numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas):
    # Creamos la matriz que llevará las cuentas del maximo beneficio posible. Se trata de una matriz, rellena a 0s por defecto.
    matrizSoluciones = np.zeros((numVacas+1, pesoMaxCamion+1), dtype=int)

    # Creamos la tabla que almacenara por cada vaca su informacion:
    ## vaca1 => (vaca1, peso1, beneficio1) => 3 datos

    matrizElementos = np.zeros((numVacas+1, 3), dtype=int)
    for i in range(numVacas+1):
        if i==0:
            pesoVaca = 0
            litrosVaca = 0
        else:
            pesoVaca = listaPesosVacas[i-1]
            litrosVaca = listaLitrosVacas[i-1]
        matrizElementos[i][0] = (i)
        matrizElementos[i][1] = pesoVaca
        matrizElementos[i][2] = litrosVaca
    
    ## NOTA: Si quieres ver la estructura y composicion de las tablas creadas, descomenta el print
    #print(matrizElementos)
    #print(matrizSoluciones) 

    return matrizElementos, matrizSoluciones
```

```python
def obtenerMaximoValor(numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas):
    
    matrizElementos, matrizSoluciones = crearMatrices (numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas)
    
    # Creamos una matriz, donde vamos a ir indicando las combinaciones que dan lugar a los beneficios(litros de leche) 
    # que se almacenarán en la matrizSoluciones
    
    matrizCombinaciones=[]
    matrizCombinaciones = [['' for col in range(pesoMaxCamion+1)] for row in range(numVacas+1)]
        
    for fila in range(1, numVacas+1):
        pElem=matrizElementos[fila,1]
        bElem=matrizElementos[fila,2]
        for columna in range(1, pesoMaxCamion+1):
            # Comprobamos si el peso de la vaca, permite incluirla en esa capacidad intermedia (peso)
            if pElem <= columna:
                #Podemos incluir ese elemento, pero tenemos que ver si su beneficio es mayor que algun anterior
                difPesos = columna - pElem
                if (bElem + matrizSoluciones[fila-1, difPesos]> matrizSoluciones[fila-1, columna]) :
                    valor = bElem + matrizSoluciones[fila-1, difPesos]
                    combinacion = 'vaca' + str(matrizElementos[fila,0]) + '+' + matrizCombinaciones[fila-1][difPesos]
                else:
                    valor= matrizSoluciones[fila-1, columna]
                    combinacion = str(matrizCombinaciones[fila-1][columna])
            else:
                # si no podemos incluir el elemento, ponemos el anterior para esa capacidad intermedia
                valor = matrizSoluciones[fila-1, columna]
                combinacion = str(matrizCombinaciones[fila-1][columna])

            matrizSoluciones[fila, columna] = valor
            matrizCombinaciones[fila][columna]= combinacion
    
    # Devolvemos como resultado:
    #  la tabla con los mejores benecificios/capacidad
    #  la tabla con las combinaciones de vacas que permiten cada uno de los beneficios de la tabla anterior

    ## NOTA: Si quieres ver la estructura y composicion de las tablas con los resultados, descomenta el print   
    #print("Matriz de Soluciones => ")
    #print(matrizSoluciones)
    #print("Matriz de Combinaciones => ")
    #print(matrizCombinaciones)
    
    # Buscamos de la matriz anterior, el maximo beneficio calculado e indicamos que combinación o combinaciones de vacas 
    # te darían ese beneficio maximo

    rdoValorMax = np.amax(matrizSoluciones)
    mensajeFinal = ''
    if rdoValorMax>0:
        mensajeFinal = "El beneficio máximo calculado es de => "  + str(rdoValorMax) + ' litros'
        rdoListaCombinaciones = np.where(matrizSoluciones == np.amax(matrizSoluciones))
        rdoCombinacionesPosibles = ""
        for j in range(len(rdoListaCombinaciones[0])):
            a = rdoListaCombinaciones[0][j]
            b = rdoListaCombinaciones[1][j]
            combinacion = matrizCombinaciones[a][b]
            ultimoCaracter = combinacion[len(combinacion)-1]
            if (ultimoCaracter == "+"):
                combinacion = combinacion[:len(combinacion)-1]
            if not combinacion in rdoCombinacionesPosibles:
                if rdoCombinacionesPosibles=="":
                    rdoCombinacionesPosibles = combinacion
                else:
                    rdoCombinacionesPosibles = rdoCombinacionesPosibles + " ó " + combinacion
        mensajeFinal = mensajeFinal + '\n' + "La/s combinacion/es de vacas que te dará/n ese beneficio es => " + rdoCombinacionesPosibles
    else:
        mensajeFinal = "No hay posibilidades, parece que todas las vacas exceden el peso máximo del camión."
    
    return mensajeFinal
```

```python
# Aqui empieza la llamada

datosOK, mesajeFinal = comprobarDatosEntrada(numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas)
if datosOK:
    mesajeFinal = obtenerMaximoValor(numVacas, pesoMaxCamion, listaPesosVacas, listaLitrosVacas)

print(mesajeFinal)
```
